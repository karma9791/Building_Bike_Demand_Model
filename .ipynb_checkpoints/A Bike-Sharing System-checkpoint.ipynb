## Dummy Variables Encoding

```python
# Get the dummy variables for the feature 'season' and store it in a new variable - 'season_dummy_var'
#drop_first drops the 1st variable as we know for k values we need k-1 data.

season_dummy_var = pd.get_dummies(bike['season'], drop_first=True)
season_dummy_var.head()
```

```python
# Get the dummy variables for the feature 'weathersit' and store it in a new variable - 'weathersit_dummy_var'

weathersit_dummy_var = pd.get_dummies(bike['weathersit'], drop_first=True)
weathersit_dummy_var.head()
```

```python
# Get the dummy variables for the feature 'mnth' and store it in a new variable - 'mnth_dummy_var'
mnth_dummy_var = pd.get_dummies(bike['mnth'], drop_first=True)
mnth_dummy_var.head()
```

```python
# Get the dummy variables for the feature 'weekday' and store it in a new variable - 'weekday_dummy_var'

weekday_dummy_var = pd.get_dummies(bike['weekday'], drop_first=True)
weekday_dummy_var.head()
```

```python
#Add the dummy variables to the original bike dataset

bike = pd.concat([bike, season_dummy_var, weathersit_dummy_var, mnth_dummy_var, weekday_dummy_var], axis=1)
bike.head()
```

```python
bike.shape
```

```python
# drop these columns whose dummies are created.
drop_variables = ['season', 'mnth', 'weekday', 'weathersit']
bike.drop(drop_variables, inplace=True, axis=1)
```

```python
bike.shape
```

```python
bike.head()
```

## Training the Model i.e. dividing into X and Y sets for model building

```python
y_train = bike_train.pop('cnt')
X_train = bike_train ## assigning all features varaibles to X_train
```

```python
# importing libraries
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()

#fitting the data
lm.fit(X_train,y_train)

# Initializing RFE:
rfe=RFE(lm,n_features_to_select=10)

# Fitting RFE to the trainning data:
rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# selecting the selected variable via RFE in col list

col = X_train.columns[rfe.support_]
print(col)
```

```python
# checking which columns has been rejected

X_train.columns[~rfe.support_]
```

```python
# Generic function to calculate VIF of variables

# Calculate the VIFs for the new model
def calculate_VIF(col,X_train):
    vif = bike.DataFrame()
    vif['Features'] = col
    vif['VIF'] = [variance_inflation_factor(X_train[col].values, i) for i in range(X_train[col].shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    display(vif)
```

```python
# dataframe with RFE selected variables

X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the 1st linear model
```
